### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]

scatter_x = np.random.rand(50)
scatter_y = np.random.rand(50)

pie_labels = ['Apple', 'Banana', 'Cherry', 'Date']
pie_sizes = [30, 40, 20, 10]

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Line chart
axs[0, 0].plot(x, y, color='b', linestyle='-', linewidth=2)
axs[0, 0].set_title('Line Chart')

# Bar chart
axs[0, 1].bar(categories, values, color='g')
axs[0, 1].set_title('Bar Chart')

# Scatter plot
axs[1, 0].scatter(scatter_x, scatter_y, color='r', alpha=0.6)
axs[1, 0].set_title('Scatter Plot')

# Pie chart
axs[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', colors=['blue', 'yellow', 'red', 'brown'])
axs[1, 1].set_title('Pie Chart')

plt.tight_layout()
plt.show()
```
